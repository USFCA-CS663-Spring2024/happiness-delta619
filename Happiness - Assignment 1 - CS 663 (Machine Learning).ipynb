## i want to see more countries covering a longer time span


```python
afg = df[df['Country name'] == 'Afghanistan'][['year', 'Life Ladder']]

bgd = df[df['Country name'] == 'Bangladesh'][['year', 'Life Ladder']]

fig, ax = plt.subplots()

ax.plot(afg['year'], afg['Life Ladder'], label='Afghanistan')

ax.plot(bgd['year'], bgd['Life Ladder'], label='Bangladesh')

ax.set_title('Declining Life Ladder Scores')

ax.set_xlabel('Year')

ax.set_ylabel('Life Ladder Score')

ax.legend()

plt.show()

```

```python
fig, ax = plt.subplots() 

countries = ['Afghanistan', 'Australia', 'Germany', 'Mexico']
for country in countries:
    country_df = df[df['Country name'] == country]
    ax.plot(country_df['year'], country_df['Negative affect'], label=country)

ax.set_xlabel('Year')
ax.set_ylabel('Negative Affect Score')
ax.legend(loc='upper left')
ax.set_title('Increasing Negative Affect in Some Countries')

plt.tight_layout()  
plt.show()

# Australia and Germany remain more steady
```

```python
china = df[(df['Country name'] == 'China')] 
costa_rica = df[(df['Country name'] == 'Costa Rica')]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(china['year'], china['Log GDP per capita'], 'ro-', label='China')  
ax1.plot(costa_rica['year'], costa_rica['Log GDP per capita'], 'bo-', label='Costa Rica')
ax1.set_title('GDP Per Capita')
ax1.legend()

ax2.plot(china['year'], china['Life Ladder'], 'ro-', label='China')
ax2.plot(costa_rica['year'], costa_rica['Life Ladder'], 'bo-', label='Costa Rica')  
ax2.set_title('Life Ladder Score')
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Filter data for the year 2022
df_2022 = df[df['year'] == 2022]
df_2022 = df_2022.sort_values(by='Life Ladder', ascending=False)
df_2022_top5 = df_2022[:15]

# Filter data for the year 2013
df_2013 = df[df['year'] == 2013]
df_2013 = df_2013.sort_values(by='Life Ladder', ascending=False)
df_2013_top5 = df_2013[:15]

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(18, 6))

sns.barplot(x='Life Ladder', y='Country name', data=df_2022_top5, palette='viridis', ax=axes[0])
axes[0].set_title('Top Happiest Countries in 2022')
axes[0].set_xlabel('Happiness Score (Life Ladder)')
axes[0].set_ylabel('Country')

sns.barplot(x='Life Ladder', y='Country name', data=df_2013_top5, palette='viridis', ax=axes[1])
axes[1].set_title('Top Happiest Countries in 2013')
axes[1].set_xlabel('Happiness Score (Life Ladder)')
axes[1].set_ylabel('Country')

plt.tight_layout()

plt.show()

```

```python
import seaborn as sns
import matplotlib.pyplot as plt


df_all_time = df.sort_values(by='Life Ladder', ascending=False)
df_all_time_top5 = df_all_time[:20]

plt.figure(figsize=(12, 6))
sns.barplot(x='Life Ladder', y='Country name', data=df_all_time_top5, palette='pastel')

plt.xlabel('Average Happiness Score (Life Ladder)')
plt.ylabel('Country')
plt.title('Happiest Countries of All Time')

plt.tight_layout()
plt.show()
```

```python
df = pd.read_excel('DataForTable2.1WHR2023.xls')
df
```

```python
import matplotlib.pyplot as plt

df['start_year'] = df.groupby('Country name')['year'].transform('min')
df['end_year'] = df.groupby('Country name')['year'].transform('max')
df['increase_in_happiness'] = df.groupby('Country name')['Life Ladder'].transform(lambda x: x.iloc[-1] - x.iloc[0])

result_df = df[['Country name', 'start_year', 'end_year', 'increase_in_happiness']].drop_duplicates()

result_df = result_df.sort_values(by='increase_in_happiness', ascending=False)


most_increase = result_df.head(5)
most_decrease = result_df.tail(5)

plt.figure(figsize=(12, 8))
plt.bar(most_increase['Country name'], most_increase['increase_in_happiness'], color='green')
plt.xlabel('Country')
plt.ylabel('Increase in Happiness')
plt.title('Top 5 Countries with the Most Increase in Happiness from starting occurence to 2022')
plt.tight_layout()

plt.show()

plt.figure(figsize=(12, 8))
plt.bar(most_decrease['Country name'], most_decrease['increase_in_happiness'], color='red')
plt.xlabel('Country')
plt.ylabel('Decrease in Happiness')
plt.title('Top 5 Countries with the Most Decrease in Happiness from starting occurence to 2022')
plt.tight_layout()

plt.show()

result_df

```
